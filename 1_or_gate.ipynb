```python
from qiskit import QuantumCircuit
qc = QuantumCircuit(3,3)

qc.h([0,1])

qc.measure(0,0)
with qc.if_test((0,1)):
  qc.x(2)
  qc.measure(2,2)

qc.measure(1,1)
with qc.if_test((1,1)):
  with qc.if_test((2,0)):
    qc.x(2)

qc.measure(2,2)
qc.draw()
```
