```python
from qiskit import QuantumCircuit
qc = QuantumCircuit(3,1) #3 qubits and 1 classical bit
qc.h([0,1]) #put control qubits into superposition
qc.ccx(0,1,2) #apply CCX gate to entangle/control target qubit based on q0 and q1 superpositions
qc.measure(2,0) #measure target qubit into classical bit
qc.draw() #draw circuit
```

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram


simulator = AerSimulator()
job = simulator.run(qc, shots=1024)
result = job.result()
counts = result.get_counts()
plot_histogram(counts)
```
