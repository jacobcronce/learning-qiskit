```python
from qiskit import QuantumCircuit
import numpy as np

qc = QuantumCircuit(3)
qc.h([0,1,2])

qc.x([0,2])
qc.mcp(np.pi, [0,1], 2)
qc.x([0,2])

qc.h([0,1,2])
qc.x([0,1,2])
qc.mcp(np.pi, [0,1], 2)
qc.x([0,1,2])
qc.h([0,1,2])
qc.measure_all()
```

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram


simulator = AerSimulator()
job = simulator.run(qc, shots=1024)
result = job.result()
counts = result.get_counts()
print(counts)
```
